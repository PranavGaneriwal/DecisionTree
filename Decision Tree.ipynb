```python
import pandas as pd
import numpy as np
from sklearn import tree
```

```python
data = pd.read_csv(r'C:\Users\prana\OneDrive\Documents\GlobalTerrorism\Iris.csv')
data.head()
```

```python
inputs = data.drop(['Species','Id'], axis = 1)
target = data['Species']
```

```python
inputs, target
```

### Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
inputs_train, inputs_test, target_train, target_test = train_test_split(inputs, target, 
                                                                        test_size = 0.2, 
                                                                        random_state = 42)
```

```python
model = tree.DecisionTreeClassifier(max_depth = 10)
model.fit(inputs_train, target_train)
```

```python
model.fit(inputs, target).get_params()
```

```python
target_pred = model.predict(inputs_test)
```

```python
target_pred
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(target_test, target_pred)
```

## Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (23,18))
fig = tree.plot_tree(model,
                    feature_names = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],
                    class_names = ['Iris-versicolor','Iris-setosa','Iris-virginica'],
                    filled = True)
```
